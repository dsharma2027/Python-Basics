```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Q1

```python
def histogram(df):
    df1 = df.select_dtypes(exclude=['object'])  
    for col in df1.columns:
        df1.hist(column = col)
        plt.xlabel(col , fontsize = 10)
        plt.title(col,fontsize = 15)
        plt.show()
    
histogram(df = pd.read_csv("cars.csv"))
```

```python
def graph(df):
    try:
        df1 = df.select_dtypes(exclude=['object'])  
        for col in df1.columns:
            df1.hist(column = col)
            plt.xlabel(col , fontsize = 10)
            plt.title(col,fontsize = 15)
            plt.show()

            df1.boxplot(column = col ,figsize = (6,8))
            plt.xlabel(col , fontsize = 10)
            plt.title(col,fontsize = 15)
            plt.show()


        df2 = df.select_dtypes(include =['object'])
        for i in df2.columns:
            x = pd.crosstab(df2[i] , columns  = 'count')
            x.plot(kind = 'bar' ,rot=0 ,color = 'red', figsize = [16,5])
    except:
        print("enter correct file name ")


        
    
graph(df = pd.read_csv("cars.csv"))
```
