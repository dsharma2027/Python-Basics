#### Guess the number game. Where user will get 3 attempts to guess the number between 1 and 9. If he is correct on 1st attempt he will be rewarded with 5000/- and in if correct in 2nd attempt rewarded with 1000 and if correctly guess in 3rd attempt there is no reward. In case user loose 500/- will dedcuted from his/her account.
#### The minimum account balace is needed to be 500/- if it  is less than that user can transfer from the e-Wallet

```python
import random
def random_num():
    return random.randint(1,9)
```

```python
def guess_num(accnt_balance,e_wallet):
   
    playing = True
    
    while True:
        num = random_num()      # To generate a random number
        count = 0               # Counter Reset
        print('The available balance is: Rs.', accnt_balance)
        
        # Checking if user has minimum balance to play the game
        
        if accnt_balance < 500:
            print("Insuficent Balance to Play!!!")
            
            # Checking if user has balance in e-Wallet for transfer
            
            if e_wallet == 0:
                print("Insufficent balance in wallet to transfer")
                print("You cannot continue further")
                print("Thanks for playing")
                playing = False    # Set playing counter to false, if e wallet balance and account balance is low
                break
            else:
                
                # Allow to transfer to from e-Wallet
                
                while e_wallet != 0:
                    withdrawl = int(input('Please enter the amount you want to withdrawl for wallet: '))
                    e_wallet = e_wallet - withdrawl
                    accnt_balance = accnt_balance + withdrawl
                    
                    # User needs to tranfer the amount from e-wallet till the account balance is >= 500
                    if accnt_balance < 500:
                        print('\n')
                        print('You still do not have enough money to start playing.')
                        print(f'your account balance is {accnt_balance}')
                        print(f'The available e-wallet balance is: {e_wallet}')
                        print('Withdraw more money from wallet')
                    else:
                        # Once the accnt balance is > 500 user can play the game
                        
                        print('Now, you are ready to play the game.')
                        print(f'The account balance is:{accnt_balance}')
                        print(f'The available e-wallet balance is: {e_wallet}')
                        break
        while playing:
            
            # Game Start
            # User will be asked for 1st guess
            
            try:
                guess = int(input("Guess the number: "))
                count +=1
                
            except:
                print("Ivalid Input, Please enter the number!!")
            else:
                
                if  guess != num:

                        if (count == 1) or (count == 2):
                            if guess > num:
                                print("Too High... Try Again!!!")
                            else:
                                print("Too Low... Try Again!!!")
                        else: 
                            # In the  third guess if user is not able to predict the number 
                            #below msg will be shown on screen
                            # 500 deducted from his account
                            
                            print('OOPS!!! You Lost!!!!')
                            print("You loose Rs. 500")
                            print("Your available balance is: ", accnt_balance - 500)
                            accnt_balance -= 500
                            break
                            
                else:
                    # In case if user guess is correct
                    
                    print('Cogratulation!!!!')
                    print(f'You won in {count} attempt') # Will print in how many attemp user won
                    
                    if count == 1:       # If user won on 1st attemp. Reward = 5000
                        accnt_balance += 5000
                        print("Jackpot!!! You have won: Rs", 5000)
                        print("Your available balance is: Rs", accnt_balance)
                        break
                    elif count == 2:
                        accnt_balance += 1000    # If user won on 2nd attemp. Reward = 1000
                        print("You have won: Rs", 1000)
                        print("Your available balance is: Rs", accnt_balance)
                        break
                    else:
                        
                        # If user won on 3rd attemp. No reward and No loss
                        
                        print("Your available balance is: Rs", accnt_balance)
                        break
        
        # User input of Y/N to Play again
        
        if input("Do you wish to play again Y/N: ").lower() == 'y':
            playing = True
            continue
        else:
            print("\nThanks for playing!!!")
            print(f'\nYour account balance is {accnt_balance} and e-wallet balance is {e_wallet}')
            break


                    


        
       
        
```
