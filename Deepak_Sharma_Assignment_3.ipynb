#### 2. Write a program to print all prime numbers from 1 to 300.


```python
def prime():
    
    print("Prime numbers between 1 and 300 are")
    count = 0
    for i in range(2,301):
        
        for j in range(2,i):
            if (i%j==0):
                print(f'{i} is not a prime')
                break
        else:
            print(f'{i} is prime')
            count +=1
    print("Number of prime between 1 and 300 are: ", count)
    

```

```python
prime()
```

#### 4. WAP which repeatedly reads numbers until the user enters “done”. Once “done” is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number. (Use of list is not allowed in this program)



```python
def series_sum():
    
    total = 0
    count = 0
    
    while True:
        
        num1 = input("Eneter the integer: ")
        if num1 == "Done":
            break
        
        try:
            
            num1 = float(num1)
            count += 1
            total = total + num1
            
        except:
            
            print("Invalid Entry!!! Please enter the numeric value")
    
    print("Number's Count is: ", count)
    print("The Total is: ", total)
    print("The Average is: ", round(total/count,2))
    
            
            
```

```python
series_sum()
```

#### 5.	Write another program that prompts for a list of numbers as above and at the end prints both the maximum and minimum of the numbers instead of the average. (Use of list is not allowed in this program)

```python
def series_sum1():
    
    total = 0
    count = 0
    mini = 9999999
    maxx = 0
    
    
    while True:
        
        num1 = input("Eneter the integer: ")
        if num1 == "Done":
            break
        
        try:
            
            num1 = float(num1)
            count += 1
            total = total + num1
            if num1 < mini:
                mini = num1
            elif num1 > maxx:
                maxx = num1
                
            
        except:
            
            print("Invalid Entry!!! Please enter the numeric value")
    
    print("Number's Count is: ", count)
    print("The Total is: ", total)
    print("The Average is: ", round(total/count,2))
    print("THe Maximum of the numbers is: ", maxx)
    print("The Minimum of the numbers is: ", mini)
            
            
```

#### 7.	Make an improvement to the Guess a number game. Guide the user where they are standing and limit the number of attempts to 3. For example, the game should go like this:


<b>
Guess the number: 4
Too low

Guess the number: 9
Too high

Guess the number: 5
Sorry you Lost!


```python
def guess_the_no1():
    num = random.randint(1,10)
    count = 0
    print('The GAME BEGINS!!!!')

    while True:
        guess = int(input('Guess the number: '))
        count += 1
        if  guess != num:
            if (count == 1) or (count == 2):
                if guess > num:
                    print("Too High... Try Again!!!")
                else:
                    print("Too Low... Try Again!!!")
                           
            elif count == 3:
                print('OOPS!!! You Lost!!!!')
                break
        else:
            print('You got it right! Congo!')
            print(f'You won in {count} chance')
            break
```

```python
guess_the_no1()
```
